```python
with open("postingslist_py3.pkl", "rb") as fh:
    pl = pickle.load(fh)
with open("database_py3.pkl", "rb") as fh:
    db = pickle.load(fh)
```

```python
plcopy = pl.copy()
dbcopy = db.copy()

while len(plcopy.keys()) > 20:
    smallest = 9999999
    smallkey = None

    for k, v in plcopy.items():
        if len(v) < smallest and k not in ['water', 'salt', 'kosher salt', 'black pepper', 'chicken']:
            smallest = len(v)
            smallkey = k
            

    print([smallest, smallkey])
    
    recipes_to_remove = plcopy[smallkey]
    del plcopy[smallkey]
    
    for recipe in recipes_to_remove:
        for k, v in plcopy.items():
            while recipe in v:
                plcopy[k].remove(recipe)
        
        
print(plcopy.keys())
```
